# Create artists_df

```python
# Function to find out about related artists in Spotify
def extract_related_artists(id):
    res = []
    for info in sp.artist_related_artists(id)['artists']:
        href = info['external_urls']['spotify']
        name = info['name']
        res.append((href, name))
    return res

def top_related_artists(ids):
    dict_res = collections.defaultdict(int)
    for id in ids:
        for href, name in extract_related_artists(id):
            dict_res[(href, name)] += 1
    return sorted(dict_res.items(), key=lambda k_v: k_v[1], reverse=True)
```

```python
# Create artists_df that has name, popularity and number of followers
def create_artists_df(indie_artists):
    dict_info = {'name': [], 'popularity': [], 'followers': []}
    for href, name in indie_artists:
        dict_info['name'].append(name)
        dict_info['popularity'].append(sp.artist(href)['popularity'])
        dict_info['followers'].append(sp.artist(href)['followers']['total'])
    return pd.DataFrame(dict_info)

artists_df = create_artists_df(indie_artists)
```

# Create indie_df

```python
# Two function to create dataframe that is full of information for each song
def extract_song_info(indie_artists):
    album_ids = []
    song_ids = []
    song_names = []
    song_artist_names = []
    artist_names = []

    for artist_id, artist_name in indie_artists:
        for info in sp.artist_albums(artist_id, limit=50)['items']:
            album_ids.append((info['id'], artist_name))
    
    for album_id, artist_name in album_ids:
        for info in sp.album_tracks(album_id, limit=50)['items']:
            song_ids.append(info['id'])
            song_names.append(info['name'])
            song_artist_names.append([artist_info['name'] for artist_info in info['artists']])
            artist_names.append(artist_name)

    return list(zip(song_ids, song_names, song_artist_names, artist_names))


def create_songs_df(song_info):
    list_info = []
    for id, name, song_artist_names, artist_name in song_info:
        dict_info = {}
        try:
            dict_info = sp.audio_features([id])[0]
            dict_info['name'] = name
            dict_info['song_artist_names'] = song_artist_names
            dict_info['artist_name'] = artist_name
        except:
            continue
        list_info.append(dict_info)
        
    return pd.DataFrame(list_info)
```

```python
# Create dataframe
song_info = extract_song_info(indie_artists)
indie_df = create_songs_df(song_info)
indie_df.head(5)
```

```python
# Remove songs that does not owned by artists
removed_indexes = []
for i, row in indie_df.iterrows():
    if row['artist_name'] not in row['song_artist_names']:
        removed_indexes.append(i)

indie_df = indie_df.drop(removed_indexes)
```

```python
# Add popularity score and released date of each song
indie_df['song_popularity'] = indie_df['id'].apply(lambda x: sp.track(x)['popularity'])
indie_df['released_date'] = indie_df['id'].apply(lambda x: sp.track(x)['album']['release_date'])
```

```python
# Clean cols
from datetime import datetime
indie_df['duration_m'] = round(indie_df['duration_ms']/60000, 2)

def convert_days(x):
    if len(x) == 4:
        return (2022 - int(x)) * 365
    return (datetime.now() - datetime.strptime(x, "%Y-%m-%d")).days

indie_df['published_days'] = indie_df['released_date'].apply(convert_days)
```

```python
# Add artist's name to song's name
indie_df['name'] = indie_df['name'] + ' - ' + indie_df['artist_name']
```

# Top recommended artists

```python
# Each artist has 100 recommended songs, we extract top artists from recommendation of list of indie artists
def recommended_artists_counter(ids):
    a_list = []
    for id in ids:
        for full_info in sp.recommendations(seed_artists=[id], seed_genres=None, seed_tracks=None, limit=100, country='VN')['tracks']:
            for info in full_info['artists']:
                a_list.append(info['name'])   
    return collections.Counter(a_list)
```

```python
# Show the result
ids = [href.split('/')[-1] for href, name in indie_artists]
top_recommended_artists = recommended_artists_counter(ids)
top_recommended_artists.most_common(20)
```

```python
# Save file csv
pd.DataFrame(data=top_recommended_artists.most_common(20),
             columns=['Name', 'Num_songs']).to_csv('top_recommended_artists.csv', index=False)
```
